```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('income.csv')
df.head()
```

```python
km = KMeans(n_clusters=3)
km
```

```python
y_predicted = km.fit_predict(df[['Age','Income($)']])
y_predicted
```

```python
df['Clusters'] = y_predicted
df
```

```python
df0 = df[df.Clusters==0]
df1 = df[df.Clusters==1]
df2 = df[df.Clusters==2]
plt0=plt.scatter(df0.Age,df0['Income($)'],color='b',marker='.')
plt1=plt.scatter(df1.Age,df1['Income($)'],color='r',marker='+')
plt2=plt.scatter(df2.Age,df2['Income($)'],color='g',marker='*')

plt.xlabel('Age')
plt.ylabel('Income($)')
plt.legend([plt0,plt1,plt2],['Income0','Income1','Income2'])
```

```python
scaler = MinMaxScaler()
scaler.fit(df[['Income($)']])
df['Income($)'] = scaler.transform(df[['Income($)']])

scaler.fit(df[['Age']])
df['Age'] = scaler.transform(df[['Age']])

df
```

```python
km = KMeans(n_clusters=3)
y_predicted = km.fit_predict(df[['Age','Income($)']])

df['Clusters'] = y_predicted
df
```

```python
df0 = df[df.Clusters==0]
df1 = df[df.Clusters==1]
df2 = df[df.Clusters==2]
plt0=plt.scatter(df0.Age,df0['Income($)'],color='b',marker='.')
plt1=plt.scatter(df1.Age,df1['Income($)'],color='r',marker='+')
plt2=plt.scatter(df2.Age,df2['Income($)'],color='g',marker='*')

plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color='black',marker='o')

plt.xlabel('Age')
plt.ylabel('Income($)')
plt.legend([plt0,plt1,plt2],['Income0','Income1','Income2'])
```

```python
k_rng = range(1,10)
sse = []

for k in k_rng:
    km = KMeans(n_clusters=k)
    km.fit(df[['Age','Income($)']])
    sse.append(km.inertia_)
```

```python
sse
```

```python
plt.plot(k_rng,sse)
plt.scatter(k_rng,sse)
```
