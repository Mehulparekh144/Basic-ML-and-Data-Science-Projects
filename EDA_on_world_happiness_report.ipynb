```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

# EDA on world-happiness-report.csv

```python
df1 = pd.read_csv('world-happiness-report.csv')
df1.head()
```

```python
happiest19= df1[(df1['Life Ladder'] > 7)  & (df1['year'] == 2019)]
```

```python
plt.style.use('bmh')
sns.barplot(x= happiest19['Country name'].unique() , y=happiest19['Life Ladder'].values)
plt.xticks(rotation='vertical')
```

```python
values = list(df1[df1['Life Ladder'] >= 7  ]['Country name'].value_counts())
country_name = df1[df1['Life Ladder'] >= 7  ]['Country name'].unique()
```

```python
sns.barplot(x=country_name,y=values)
plt.xticks(rotation='vertical')
```

```python
df1[df1['Country name'] == 'India']['Life Ladder']
```

```python
sns.kdeplot(df1[df1['Country name'] == 'India']['Life Ladder'] , shade=True)
plt.title("India's Life Ladder Over the years")
plt.show()
```

```python
sns.kdeplot(df1[df1['Country name'] == 'India']['Life Ladder'] , shade=True)
sns.kdeplot(df1[df1['Country name'] == 'Australia']['Life Ladder'] ,color='y', shade=True)
plt.show()
```

```python
sns.kdeplot(df1[df1['Country name'] == 'India']['Life Ladder'] , shade=True)
sns.kdeplot(df1[df1['Country name'] == 'Pakistan']['Life Ladder'] ,color='g', shade=True)
plt.show()
```

```python
sns.barplot(x=df1[df1['Country name'] == 'India']['year'],y=df1[df1['Country name'] == 'India']['Life Ladder'])
plt.show()
```

```python
df = pd.read_csv('world-happiness-report-2021.csv')
df.head()
```

```python
df.drop(['Standard error of ladder score','upperwhisker','lowerwhisker','Explained by: Log GDP per capita','Explained by: Social support','Explained by: Healthy life expectancy','Explained by: Freedom to make life choices','Explained by: Generosity','Explained by: Perceptions of corruption','Dystopia + residual'],axis=1,inplace=True)
```
