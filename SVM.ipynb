```python
import pandas as pd
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
dir(iris)
```

```python
iris['feature_names']
```

```python
df = pd.DataFrame(iris.data,columns=iris.feature_names)
df
```

```python
df['target'] = iris.target
df
```

```python
df['flower_name'] = df.target.apply(lambda x: iris.target_names[x])
df
```

```python
import matplotlib.pyplot as plt
```

```python
df0 = df[df.target == 0]
df1 = df[df.target == 1]
df2 = df[df.target == 2]
```

```python
df2.head()
```

```python

plt.scatter(df0["sepal length (cm)"],df0["sepal width (cm)"],color='g',marker='*')
plt.scatter(df1["sepal length (cm)"],df1["sepal width (cm)"],color='b',marker='+')
plt.xlabel("sepal length (cm)")
plt.ylabel("sepal width (cm)")
```

```python
plt.scatter(df0["petal length (cm)"],df0["petal width (cm)"],color='g',marker='*')
plt.scatter(df1["petal length (cm)"],df1["petal width (cm)"],color='b',marker='+')
plt.xlabel("petal length (cm)")
plt.ylabel("petal width (cm)")
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = df.drop(['target','flower_name'],axis='columns')
y = df.target
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.svm import SVC
model = SVC()
```

```python
model.fit(x_train,y_train)
```

```python
model.predict([[6.3,3.3,6.0,2.5]])
```

```python
model.score(x_test,y_test)
```

```python
from sklearn.datasets import load_digits
digits = load_digits()
```

```python
dir(digits)
```

```python
digits.target
```

```python
df = pd.DataFrame(digits.data , digits.target)
df
```

```python
df['targets'] = digits.target
df.head()
```

```python
from sklearn.model_selection import train_test_split
x = df.drop(df.targets,axis='columns')
y = df.targets

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.svm import SVC
```

```python
rbf_model = SVC(kernel='rbf')
```

```python
rbf_model.fit(x_train,y_train)
```

```python
rbf_model.score(x_test,y_test)
```

```python
modle = SVC(kernel='linear')
modle.fit(x_train,y_train)
modle.score(x_test,y_test)
```
