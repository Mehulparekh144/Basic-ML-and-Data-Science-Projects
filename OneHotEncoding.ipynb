```python
import pandas as pd
df = pd.read_csv('homeprices2.csv')
df
```

```python
dummie = pd.get_dummies(df.town)
dummie
```

```python
merged = pd.concat([df,dummie],axis = 'columns')
merged
```

```python
final = merged.drop(['town','west windsor'],axis=1)
final
```

```python
from sklearn import linear_model
model = linear_model.LinearRegression()
```

```python
x = final.drop(['price'],axis=1)
y = final[['price']]
r = model.fit(x,y)
r.predict([[2700,0,1]])
```

```python
r.predict([[3000,0,0]])
```

```python
r.score(x,y)
```
